```python
def strf_date(x):
    if pd.isnull(x): return ""
    str_x = str(int(x))
    return str_x[0:4] + '-' + str_x[4:6] + '-' + str_x[6:]
```

```python
from datetime import date
src = src.applymap(lambda x: x.strip() if type(x) == str else x)
date_rows = ['Z71_OPEN_DATE', 'Z30_PROCESS_STATUS_DATE', 'Z30_UPDATE_DATE', 'UPDATE_DATE', 'Z68_OPEN_DATE', 'Z68_ORDER_STATUS_DATE_X']
for d in date_rows:
    src[d] = src[d].astype(float)
    src[d] = src[d].apply(strf_date)
    print("row %s done" %d)
```
