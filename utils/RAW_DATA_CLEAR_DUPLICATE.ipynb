```python
src = pd.read_excel("../assets/source/NEW_REPORT.xlsx", dtype=str)
if int(src.iloc[0]['Z68_OPEN_DATE']) > int(src.iloc[-1]['Z68_OPEN_DATE']):
    src = src.iloc[::-1]
    src = src.reset_index(drop=True)
```

```python
def strf_date(x):
    if pd.isnull(x): return ""
    return x[0:4] + '-' + x[4:6] + '-' + x[6:]
```

```python
from datetime import date
src = src.applymap(lambda x: x.strip() if type(x) == str else x)
date_rows = ['Z30_PROCESS_STATUS_DATE', 'Z30_UPDATE_DATE', 'UPDATE_DATE', 'Z68_OPEN_DATE', 'Z68_ORDER_STATUS_DATE_X']
for d in date_rows:
    src[d] = src[d].apply(strf_date)
    print("row %s done" %d)
```
